```python
import numpy as np

def generateTuple(G1,k):
    if k==0:
        return [[]]
    else:
        lis=[]
        for x in G1:
            dd=generateTuple(G1,k-1)
            for s in dd:
                l=[]
                l.append(x)
                l.extend(s)
                lis.append(l)
        return lis
    
def getString(l):
    s=''
    for x in l:
        s+=str(x)+'_'
    return s

def isotest(l1,G1,l2,G2):
    for i in range(len(l1)):
        for j in range(len(l1)):
                if l1[i]==l1[j] and l2[i]!=l2[j] :
                    return False
                if l1[j] in G1[l1[i]] and l2[j] not in G2[l2[i]] :
                    return False
                if l2[i]==l2[j] and l1[i]!=l1[j] :
                    return False
                if l2[j] in G2[l2[i]] and l1[j] not in G1[l1[i]] :
                    return False
    return True

def hash1(t1,g1,t2,g2):
    lisid1=[]
    lisid2=[]
    li=[]
    for x in t1:
        bol=True
        for (y,g,i) in li:
            if isotest(y,g,x,g1):
                bol=False
                lisid1.append(i)
                break
        if bol:
            o=len(li)
            li.append((x,g1,o))
            lisid1.append(o)
    for x in t2:
        bol=True
        for (y,g,i) in li:
            if isotest(y,g,x,g2):
                bol=False
                lisid2.append(i)
                break
        if bol:
            o=len(li)
            li.append((x,g2,o))
            lisid2.append(o)
            
    return lisid1,lisid2

def mapl(l):
    di={}
    for i in range(len(l)):
        di[getString(l[i])]=i
    return di

def getCS(l,sep='_'):
    s=[str(a) for a in l]
    s.sort()
    f=''
    for c in s:
        f+=c+sep
    return f

def getCS1(l,sep='_'):
    s=[str(a) for a in l]
    f=''
    for c in s:
        f+=c+sep
    return f

def maxl(l1,l2):
    a=np.max(np.array(l1))
    b=np.max(np.array(l2))
    if a>b:
        return a
    else:
        return b
    
def N(s,i,g11):
    l=[]
    for v in g11:
        h=s.copy()
        h[i]=v
        l.append(h)
    return l
        

def hash2(t1,g1,d1,l1,t2,g2,d2,l2):
    di={}
    ll1=[]
    for s in t1:
        l=[]
        for i in range(len(t1[0])):
            c=[l1[d1[getString(g)]] for g in N(s,i,g1)]
            l.append(getCS(c))
        ss=str(l1[d1[getString(s)]])+'*'+getCS1(l,sep='*')
        if ss not in di:
            di[ss]=len(di)
        ll1.append(di[ss])
    ll2=[]
    for s in t2:
        l=[]
        for i in range(len(t2[0])):
            c=[l2[d2[getString(g)]] for g in N(s,i,g2)]
            l.append(getCS(c))
        ss=str(l2[d2[getString(s)]])+'*'+getCS1(l,sep='*')
        if ss not in di:
            di[ss]=len(di)
        ll2.append(di[ss])
    return ll1,ll2

def N1(s,v,l1,d1):
    ll=[]
    for i in range(len(s)):
        h=s.copy()
        h[i]=v
        ll.append(l1[d1[getString(h)]])
    #print('w',len(ll),ll)
    return ll

def hash3(t1,g1,d1,l1,t2,g2,d2,l2):
    di={}
    ll1=[]
    for s in t1:
        l=[]
        for v in g1:
            c=N1(s,v,l1,d1)
            #print(s)
            bb=getCS1(c)
            #print('f',bb,c)
            l.append(bb)
        ss=str(l1[d1[getString(s)]])+'*'+getCS(l,sep='*')
        #print(s,ss,l)
        if ss not in di:
            di[ss]=len(di)
        ll1.append(di[ss])
    ll2=[]
    for s in t2:
        l=[]
        for v in g2:
            c=N1(s,v,l2,d2)
            l.append(getCS1(c))
        ss=str(l2[d2[getString(s)]])+'*'+getCS(l,sep='*')
        #print(s,ss)
        if ss not in di:
            di[ss]=len(di)
        ll2.append(di[ss])
    return ll1,ll2

    

def kDimWL(k,G1,G2):
    if k<2:
        print('error')
        return -1
    maxa=-1
    t1=generateTuple(G1,k)
    t2=generateTuple(G2,k)
    d1=mapl(t1)
    d2=mapl(t2)    
    l1,l2=hash1(t1,G1,t2,G2)
    while maxa<maxl(l1,l2):
        maxa=maxl(l1,l2)
        print('i')
        if getCS(l1)!=getCS(l2):
            return False
        l1,l2=hash2(t1,G1,d1,l1,t2,G2,d2,l2)
    return True

def printa(l, t):
    print('')
    for i in range(len(l)):
        print(t[i],l[i])
    print('')

def kDimFWL(k,G1,G2):
    if k<1:
        print('error')
        return -1
    maxa=-1
    t1=generateTuple(G1,k)
    t2=generateTuple(G2,k)
    print(len(t1),len(t1))
    d1=mapl(t1)
    d2=mapl(t2)    
    l1,l2=hash1(t1,G1,t2,G2)
    while maxa<maxl(l1,l2):
        #print('i',maxa)
        maxa=maxl(l1,l2)
        print('i',maxa)
        if getCS(l1)!=getCS(l2):
            return False
        l1,l2=hash3(t1,G1,d1,l1,t2,G2,d2,l2)
    return True
    
```
